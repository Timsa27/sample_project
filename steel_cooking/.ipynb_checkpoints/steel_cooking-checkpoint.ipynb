```python
#функция построения плотности распределения
def density_distribution(data, title, xlabel,ylim=[0,0.02]):
    sns.distplot(data, hist=True)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.xlim([data[data>0].min(),data.max()])
    plt.ylim(ylim);
```

Построим графики распределения плотности объемов проволочного матрериала.
Построим в одним

```python
columns = data_wire.drop('key', axis =1).columns
for column in columns:
    count_wire = data_wire[column].count()
    plt.figure(figsize=(15,5))
    density_distribution(data_wire[column].fillna(0).astype(int), 
                         'Плотность распределения {}, количество добавлений {}'.format(column, count_wire), 
                         'Объем проволчного материала',
                         ylim = [0,0.02])
    
```

```python
density_distribution(data_wire.drop('key', axis=1).sum(axis=1),
                     'Плотность распределения суммарного объема проволочных материалов', 
                     'Объем проволочных материалов')
print('Средний суммарный объем проволочных материалов по партиям:', data_wire.drop('key', axis=1).sum(axis=1).mean())
```

```python
# создадим функцию, которая подберет лучшие гиперпараметры с помощью кросс-валидации
def train_model(model, params):
    grid_model = GridSearchCV(model, params, scoring='neg_mean_absolute_error', 
                              cv=6,
                              refit=True, n_jobs=-1)
    
    grid_model.fit(features_train.drop(['Bulk 3','Bulk 4','Wire_other'],axis=1), target_train)

    return (grid_model.best_params_,grid_model.best_estimator_,grid_model.best_score_)
```

```python
#Создадим функцию, которая принимает модели с гиперпараметрами и возвращает датафрейм с лучшими параметрами
def eval_models(models):
    model_results = [
        [model.__class__.__name__, *train_model(model, params)] for model, params in models
    ]
    
    columns=['Model Name', 'best_params','model', 'score']
    result= pd.DataFrame(model_results, columns=columns).set_index('Model Name').sort_values('score', ascending=False)
    
    
    display(result)
    
    return model_results
```

# Отчет по проекту:
1. Первой задачей по проекту я поставил изучить технологический процесс и понять приницип изменения температуры взависимости от факторов. Задача была решена и выстроена теоретическая модель:
    - Основой для понимания стала формула
    
    $T=T0 + c*A/m$
    , где 
        - с - удельная теплоемкость
        - А - Работа или переданная энергия(при подогреве)
        - m - масса
    - Температура отражает внутреннюю энергию, поэтому если смешиваются два материала разной температурой, то суммарная температура будет средняя температура между двумя материалами, умноженная на коэф.( части массы материала от общей массы), т.е. чем больше масса материала, тем больше вклад его температура внесет в конечную температура, 
    $T = T1*k1 + T2*k2$
    
    $k1 = m1/(m1+m2)$
    
    $k2 = m2/(m1+m2)$
    
    т.е. связь здесь не линейная, если учесть тот факт, что масса сплава в ковше намного больше, чем масса примесей, то можно принять связь линейной, и т.к. за счет примесей сплав будет остывать, то корреляция будет отрицательной
    - влияние газа на температуру смоделировать сложно, но думаю что с определенными допусками, можно предположить, что тоже линейная, и т.к. за счет примесей сплав будет остывать, то корреляция будет отрицательной

  2. На основании сколько начальной температуры и целевой (рекомендуемая температура для легирования), предполагаем насколько мы должны подогреть. Т.е. чем меньше первоначальная температура, тем больше требуется подогрев.
