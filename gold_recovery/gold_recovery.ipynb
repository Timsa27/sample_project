## Проверим, что эффективность обогащения рассчитана правильно. 
Вычислим её на обучающей выборке для признака rougher.output.recovery. Найдем MAE между вашими расчётами и значением признака. Выводы.

MAE (англ. mean absolute error) 
на вход подадим столбец `rougher.output.recovery` из `gold_recovery_train` и добавим столбец `calculated_rougher.output.recovery` по формуле:

recovery = (C*(F-T))/(F*(C-T))*100 (%)
где:
- С - rougher.output.concentrate_au (доля золота в концентрате после флотации)
- F - rougher.input.feed_au (доля золота в сырье до флотации)
- T - rougher.output.tail_au (доля золота в хвосте после флотации)



```python
# Создадим функцию по расчету recovery, которая принимает на вход таблицу с данными, 
# колонки с концентрацией золота на выходе(C), на входе (F) и в хвосте (T)
# и добавляет стобец 'calculated_rougher.output.recovery' с расчитанным recovery

def recovery_calculate(data, output_concentrate_au, input_feedconcentrate_au, output_tail_au ):
    # зададим переменные 
    C = data[output_concentrate_au]
    F = data[input_feedconcentrate_au]
    T = data[output_tail_au]
    #расчитаем recovery
    recovery_data = (C*(F-T))/(F*(C-T))*100
    return recovery_data 
```

```python

gold_recovery_train['calculated_rougher.output.recovery'] = (
    recovery_calculate(gold_recovery_train,
                       'rougher.output.concentrate_au',
                       'rougher.input.feed_au', 
                       'rougher.output.tail_au')
)

# Посмотрим на 5 строк первичных и рассчитанных значений recovery
gold_recovery_train[['rougher.output.recovery','calculated_rougher.output.recovery']].head()
```

Пропуски есть как в данных значениях  recovery, так и в расчитанных.
Посмотрим на строки с пропущенными значениями в рассчитаных recovery 

```python
(gold_recovery_train[gold_recovery_train['calculated_rougher.output.recovery']
                     .isnull()]
 [['rougher.output.concentrate_au',
   'rougher.input.feed_au',
   'rougher.output.tail_au',
   'calculated_rougher.output.recovery',
   'rougher.output.recovery']]
)
```

Пропуски в рассчитанных данных определяются формулой, что в знаменателе 0 ( т.е. доля золота в сырье до флотации равна 0), судя по тому, что в отвальных хвостах есть содержание золота, то 0 это либо заполненый пропуск, либо ошибка.

Для расчета MAE между  расчитанным recovery `calculated_rougher.output.recovery` и данным recovery `rougher.output.recovery` будем использовать не нулевые значения признаков

```python
# рассчитаем MAE, удалив строки с пропусками в произнаках
recovery_drop_nun = gold_recovery_train[['rougher.output.recovery','calculated_rougher.output.recovery']].dropna()
# Напечатем среденее значение recovery, чтобы оценить ошибку
print('Среднее значение recovery', recovery_drop_nun['rougher.output.recovery'].mean())


print('MAE', mean_absolute_error(recovery_drop_nun['rougher.output.recovery'],recovery_drop_nun['calculated_rougher.output.recovery']))
```

Среднее абсолютное отклонее между признаками `rougher.output.recovery` и расчитанным recovery `calculated_rougher.output.recovery` достаточно мало ( 9.4e-15), делаем вывод, что в  recovery посчитано верно.

## Функция для вычисления итоговой sMAPE.

```python
# напишем функцию sMAPE, которая принимает на вход Series из ответов и список из предсказаний, а выводит sMAPE
def sMAPE_calculate(target,predict):
    
    
    return 100 * 2 * ((target - predict).abs() / (target.abs() + predict.abs())).mean()

# напишем функцию sMAPE_calculate, которая принимает на вход Dataframe из двух колонкох ответов по черновому этапу и финальному и список списков, в каждом из которого будет два 
# ответа по черновому этапу и финальному
def sMAPE_final(target,predict):
    
    predict = pd.DataFrame(predict, columns = target.columns, index = target.index)
    # определим столбцы
    predict_rougher = predict['rougher.output.recovery']
    predict_final = predict['final.output.recovery']    
    target_rougher = target['rougher.output.recovery']
    target_final = target['final.output.recovery']    
    
    sMAPE_rougher = sMAPE_calculate(target_rougher,predict_rougher)
    sMAPE_final = sMAPE_calculate(target_final,predict_final)
    
    return sMAPE_rougher*0.25 + sMAPE_final*0.75
```
