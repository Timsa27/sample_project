### Разобъем данные на обучающую и валидационную выборки в соотношении 75:25.

```python
# выделим целевой параметр по 0 региону
target = geo_data_1['product']
# выделим признаки
features = geo_data_1.drop(['product','id'], axis=1)
```

```python
features_train, features_valid, target_train, target_valid = train_test_split(
    features, target, test_size=0.25, random_state=12345)
```

```python
# нормализуем признаки (у нас все признаки - числовые) 
scaler = StandardScaler()

scaler.fit(features_train)
features_train = scaler.transform(features_train)
features_valid = scaler.transform(features_valid)
```

### Обучите модель и сделайте предсказания на валидационной выборке.

```python
# по условию задачи дано, что у нас работает только линейная регрессия
# обучим линейную регрессию 
model = LinearRegression()
model.fit(features_train, target_train)
```

### Сохраните предсказания и правильные ответы на валидационной выборке.

```python
# построим предсказания на валидационных данных 
predicted_valid = model.predict(features_valid) 
```

### Напечатаем на экране средний запас предсказанного сырья и RMSE модели.

```python

#  т.к. целевой признак product не имеет выбросов, то за среднее будем брать среднее арифметическое
mean_predicted_valid = predicted_valid.mean()

print("Средний запас предсказанного сырья", mean_predicted_valid)

# посчитаем качество модели на валидационной выборке

mse = mean_squared_error(target_valid, predicted_valid)
print("RMSE =", mse ** 0.5)
```

### Проанализируйте результаты.

```python
# посмотрим на RMSE для константной модели
mean_target = target_train.mean()
predicted_valid = pd.Series(mean_target, index=target_valid.index)# < напишите код здесь >
mse = mean_squared_error(target_valid, predicted_valid)
print("Mean")
print("RMSE =", mse ** 0.5)
```

Мы проверили адекватность нашей модели, сравнив среднеквадратическую ошибку ее предсказаний с предсказаниями константной модели:
RMSE линейной регрессии : 37.6
RMSE константной модели(среднего) : 44.3

Модель проходит тест на адекватность.

```python
def mean_product_RMSE(geo_data):
    
    # выделим целевой параметр по региону
    target = geo_data['product']
    # выделим признаки
    features = geo_data.drop(['product','id'], axis=1)
    
    features_train, features_valid, target_train, target_valid = train_test_split(
    features, target, test_size=0.25, random_state=12345)
    
    # нормализуем признаки (у нас все признаки - числовые) 
    scaler = StandardScaler()

    scaler.fit(features_train)
    features_train = scaler.transform(features_train)
    features_valid = scaler.transform(features_valid)

    # обучим линейную регрессию 
    model = LinearRegression()
    model.fit(features_train, target_train)
    
    # построим предсказания на валидационных данных 
    predicted_valid = model.predict(features_valid) 

    #  считаем среднее
    mean_predicted_valid = predicted_valid.mean()
    
    
    # запишем предсказанный продуктк в виде Serias, индекс будем использовать от  target_valid
    predicted_valid=pd.Series(predicted_valid, target_valid.index)

      #Средний запас предсказанного сырья

    # посчитаем качество модели на валидационной выборке

    mse = mean_squared_error(target_valid, predicted_valid)
    RMSE = mse ** 0.5
    return [mean_predicted_valid, RMSE, predicted_valid,target_valid] # RMSE среднеквадратическая ошибка
```

```python
# созраним Средний запас предсказанного сырья и RMSE среднеквадратическую ошибку для каждого региона
mean_predicted_0, RMSE_0, predicted_valid_0,target_valid_0 = mean_product_RMSE(geo_data_0)
mean_predicted_1, RMSE_1, predicted_valid_1,target_valid_1 = mean_product_RMSE(geo_data_1)
mean_predicted_2, RMSE_2, predicted_valid_2,target_valid_2 = mean_product_RMSE(geo_data_2)
data_region_mean_rmse= {'mean_predicted' : [mean_predicted_0, mean_predicted_1, mean_predicted_2],
        'RMSE_' : [RMSE_0, RMSE_1, RMSE_2]}
region_mean_rmse = pd.DataFrame(data_region_mean_rmse)
region_mean_rmse
```

## Расчёт прибыли и рисков для каждого региона 

### Примените технику Bootstrap с 1000 выборок, чтобы найти распределение прибыли.


Напишем функцию для бустинга по регионам, которая принимает на вход результат предсказаний `predicted_valid`, и целевой принзнак `target_valid`,  а выводит Series в виде переменной `profit` из возможных значений прибыли в **млрд. рублей**

```python
def profit_region(predicted_valid, target_valid):
    
    state = np.random.RandomState(12345)
    
    #создадим список profit, в котором, будем добавлять значение прибыли
    profit = []
    for i in range(1000):
        #создадим выборку целевого признака из target_valid в заданом количестве (по условию 500)
        target_subsample = target_valid.sample(n=500, replace=True, random_state=state)
        
        #создадим переменную, в которой сохраним пресказания продукта в отобранных скважинах
        predicted_subsample = predicted_valid[target_subsample.index]
    
        #  отсортируем предсказанный продукт по убыванию
        predicted_valid_sort=predicted_subsample.sort_values(ascending=False)

        # выберем значений продукта 200 скважин, в которых было предсказно больше всего продукта
        selected = target_subsample[predicted_valid_sort.index][:200]
    
        # посчитаем доход со всех отобранных (по условию 200) скажин в млрд рублей
        sum_product = selected.sum()*BAR_BILLION
        

        # запишим прибыль с учетом, что на разработку 200 скважин потратили 10 млрд р.
        profit.append(sum_product-10)

    profit = pd.Series(profit)
    return profit
```

```python
#  сохраним 1000 возможных значений прибыли по 3 регионам 
profit_0 = profit_region(predicted_valid_0, target_valid_0)
profit_1 = profit_region(predicted_valid_1, target_valid_1)
profit_2 = profit_region(predicted_valid_2, target_valid_2)
```

Напишем функцию, которая принимает на вход список возможной прибыли, а выводит:
    - вероятность убытков;
    - среднее значение прибыли одной скважины в регоине, млрд.р.
    - доверительный интревал прибыли на одной скважине

```python
def profit_lossprob_mean_confint(profit, region):
    return [round((profit < 0).mean() * 100,2),
    round(profit.mean(), 2),
    round(profit.quantile(0.025),2),
    round(profit.quantile(0.975),2)]
```

Составим таблицу в котором отобразим вероятность убытков, Ср. значение продукта и доверительный интревал для каждого региона
