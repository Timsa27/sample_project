```python
#  для наглядности построим датафрейм из столбцов основного файла данных, и добавим столбцы, в которых 
# выделим количество не пропущенных значений в стоблце, количество пропущенных и процент пропущенных значений
# и столбец, который указывает тип данных в столбце 
# создадим функцию, которая будет это выполнять
def pass_stat(data):
    pass_data = pd.DataFrame({'columns': list(data.columns)})
    i = 0
    for column in pass_data['columns']:
        count_1 = data[column].count()
        pass_data.loc[i,'count'] = count_1
        pass_data.loc[i,'pass_count'] = len(data) - count_1
        pass_data.loc[i,'pass_count_%'] = (len(data) - count_1)/len(data)*100
        pass_data.loc[i,'type'] = data[column].dtypes
        i += 1

    pass_data = pass_data.query('pass_count>0').sort_values('pass_count').reset_index()
    print(pass_data)
    
pass_stat(data)
```

Посмотрим на следующую колонку по рейтингу пропущенных значений это parks_nearest (растояние до ближайшего парка) по описанию максимум это 3190 м , а минимум это 1 м. . так же есть столбец, который указывает на количество парков   в пределах 3 км, parks_around3000, если в это ячейке 0, то таким пропущенным значениям можно присвоить значения "много" больше 3 км например 10 км (что будет означать, что парка рядом нет), про предыдущим выведенным строкам, такие строки есть можно сразу фильтровать и менять, если в пределах 3 км есть парк, то можно расстояние заменить на среднее значение которое будер меньше 3 км 



```python
# запишим среднее расстояние до парка, если он есть в пределах 3 км в "parks_nearest_mean_less3000"
parks_nearest_mean_less3000 = data.query('parks_around3000 > 0')['parks_nearest'].mean() 


# если парков нет до 3 км, то сделаем такое растояние 10000
data.loc[data['parks_around3000'] == 0, 'parks_nearest'] = 10000
# если парк есть до 3 км, то присвоим расстоянию среднее значение.
data.loc[data['parks_around3000'] > 0,'parks_nearest'] = data.loc[data['parks_around3000'] > 0,'parks_nearest'].fillna(parks_nearest_mean_less3000)
```

```python
# далее заполним пропуски, в столбце с растоянием до водоема способом аналогичным расстоянию до парка

# среднее расстояние до водоема, если он есть в пределах 3 км
ponds_nearest_mean_less3000 = data.query('ponds_around3000 > 0')['ponds_nearest'].mean() 


# если водоема нет до 3 км, то сделаем такое растояние 10000
data.loc[data['ponds_around3000'] == 0, 'ponds_nearest'] = 10000
# если водоем есть до 3 км, то присвоим расстоянию среднее значение.
data.loc[data['ponds_around3000'] > 0,'ponds_nearest'] = data.loc[data['ponds_around3000'] > 0,'ponds_nearest'].fillna(ponds_nearest_mean_less3000)
pass_stat(data)
```

```python
# СПБ нет в списке и просматриваются: поселок, село, деревня, пгт и не большие города.
# проверим во всех ли ячейках в данных городах пропуски в столбцах с расстоянием
# создатим таблицу, в которой будет список вышеуказанных населенных пунктов
# и количетсво ненулевых значений по столбцам расстояний
pass_data_not_spb = pd.DataFrame ({'columns': data[data['parks_around3000'].isnull()]['locality_name'].unique()})
m = 0

for row in pass_data_not_spb['columns']:
    pass_data_not_spb.loc[m,'airports_nearest_count'] = data.query('locality_name == "row"')['airports_nearest'].count()
    pass_data_not_spb.loc[m,'cityCenters_nearest_count'] = data.query('locality_name == "row"')['cityCenters_nearest'].count()
    pass_data_not_spb.loc[m,'parks_around3000_count'] = data.query('locality_name == "row"')['parks_around3000'].count()
    pass_data_not_spb.loc[m,'parks_nearest_count'] = data.query('locality_name == "row"')['parks_nearest'].count()
    pass_data_not_spb.loc[m,'ponds_around3000_count'] = data.query('locality_name == "row"')['ponds_around3000'].count()
    pass_data_not_spb.loc[m,'ponds_nearest_count'] = data.query('locality_name == "row"')['ponds_nearest'].count()
    m += 1

#pass_data_not_spb = pass_data_not_spb..sort_values('pass_count').reset_index()
pass_data_not_spb.sort_values('airports_nearest_count', ascending = False).head()
```

```python
# также посмотрим на корреляцию жилой и общей площади
data['living_area'].corr(data['total_area'])
```

```python
# возможно у сгруппированных по количеству комнат будет получше корреляция
# создадим дата фрейм в котором отобразим количество комнат и коэффициенты корреляции площади кухни и общей площади
# и жилой площади к площади кухни

rooms_count = pd.DataFrame({'rooms_count': data['rooms'].unique()})
# напишим функции которая принимает количество комнат и возвращать коэф коррялиции м/у площади кухни и общей площади 
def rooms_kitchen_total_area_corr(rooms_count):
    
    return data.query('rooms == @rooms_count')['kitchen_area'].corr(data.query('rooms == @rooms_count')['total_area'])
# напишим функции которая принимает количество комнат и возвращать коэф коррялиции м/у жилой площадью и общей площади 
def rooms_living_total_area_corr(rooms_count):
    
    return data.query('rooms == @rooms_count')['living_area'].corr(data.query('rooms == @rooms_count')['total_area'])
#  добавим соответствующие столбцы
rooms_count['corr_living_total_area']=rooms_count['rooms_count'].apply(rooms_kitchen_total_area_corr)
rooms_count['corr_kitchen_total_area']=rooms_count['rooms_count'].apply(rooms_living_total_area_corr)

# добавим столбец с количеством объявлейний по кождому количество комнат, 
# чтобы оценить по какому количество строилась корреляция
 
rooms_count.join(data['rooms'].value_counts(), 'rooms_count')

#  посмотрим на результат
```

```python
#  по таблице видно, что у сгрупированных по количеству комнат объявлений корреляция выше
# значит можно считать отношение площади кухни к общей площади и по среднему коэффициенту в группах по кол-ву комнат
# вычислять пропущенные значения
# добавим в исходный data столбец с отношением площади кухни к общей площади

data['kitchen_area_total_area'] = data['kitchen_area']/data['total_area']

# созададим ф-ю, которая принимает количество комнат возвращает среднее значение отношения площади кухни 
# к общей площади

def kitchen_area_total_area (rooms_count):
    return data.query('rooms == @rooms_count')['kitchen_area_total_area'].mean()
   
# там где корреляция выше 0.5 посчитаем отношение внутри группы
rooms_count['kitchen_area_total_area'] = (
    rooms_count.query('corr_kitchen_total_area > 0.5')['rooms_count']
    .apply(kitchen_area_total_area)
)

# там где корреляция ниже 0.5 посчитаем отношение, присвоив среднее знание по всему столбцу
rooms_count['kitchen_area_total_area'] = (
    rooms_count['kitchen_area_total_area']
    .fillna(data['kitchen_area_total_area']
            .mean())
)

rooms_count
```

```python
# проведем аналогичные действия с жилой площадью
# добавим в исходный data столбец с отношением жилой площади к общей площади

data['living_area_total_area'] = data['living_area']/data['total_area']

# созададим ф-ю, которая принимает количество комнат возвращает среднее значение отношения жилой площади  
# к общей площади

def living_area_total_area (rooms_count):
    return data.query('rooms == @rooms_count')['living_area_total_area'].mean()
   
# там где корреляция выше 0.5 посчитаем отношение внутри группы
rooms_count['living_area_total_area'] = (
    rooms_count.query('corr_living_total_area > 0.5')['rooms_count']
    .apply(living_area_total_area)
)

# там где корреляция ниже 0.5 посчитаем отношение присвоим среднее знание по всему столбцу
rooms_count['living_area_total_area'] = (
    rooms_count['living_area_total_area']
    .fillna(data['living_area_total_area']
            .mean())
)

rooms_count
```

```python
# добавим столцы в data с кооффициентами, для этого переименуем rooms_count в rooms

rooms_count.columns = ['rooms', 'corr_living_total_area', 'corr_kitchen_total_area', 'kitchen_area_total_area', 'living_area_total_area']

data = data.merge(rooms_count, on = 'rooms', how = 'left')
```

```python
# заменим пропущенные значения площади кухни произведение общей площади на коэф, в зависимости от кол-ва комнат
data['kitchen_area'] = data['kitchen_area'].fillna(data['total_area']*data['kitchen_area_total_area_y'])
data['living_area'] = data['living_area'].fillna(data['total_area']*data['living_area_total_area_y'])
pass_stat(data)
```
